```python
from langchain_openai import OpenAIEmbeddings
```

```python
embed_agent = OpenAIEmbeddings()

docs_embeddings = embed_agent.embed_documents([
    "Hi there!",
        "Oh, hello!",
        "What's your name?",
        "My friends call me World",
        "Hello World!"
])

len(docs_embeddings), len(docs_embeddings[0])
```

```python
embedded_query = embed_agent.embed_query("What was the name mentioned in the conversation?")
embedded_query[:5]
```

## Using Hugging Face Models

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embed_agent = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
sentences = ["This is an example sentence", "Each sentence is converted"]
result = embed_agent.embed_documents(sentences)
```
